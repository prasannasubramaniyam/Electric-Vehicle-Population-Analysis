#1. Univariate Analysis (3–5 Visuals)

**Histogram - Electric Range Distribution**

```python
#1. shows how electric vehicle ranges are distributed.
plt.hist(electrical_vehicle_population['Electric Range'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Electric Range')
plt.ylabel('Frequency')
plt.title('Distribution of Electric Range')
plt.show()
```

**Interpretation Of Histogram - Electric Range Distribution**

1️⃣ Distribution Pattern
The histogram shows how many vehicles fall into different electric range intervals (for example: 0–50 miles, 50–100 miles, etc.).

2️⃣ Most Common Range
The tallest bars indicate the range values where most electric vehicles are concentrated.

3️⃣ Low Range Vehicles
Some vehicles appear in the lower range categories, which are usually older models or plug-in hybrid vehicles (PHEVs).

4️⃣ High Range Vehicles
Vehicles with higher electric ranges represent newer or more advanced EV models.

5️⃣ Trend Observation
The distribution may show that most vehicles have moderate electric ranges, while only a smaller number reach very high ranges.

**Bar Chart Top EV Manufactures**

```python
#2. shows the most common EV brands
brand_counts = electrical_vehicle_population['Make'].value_counts().head(10)
plt.figure(figsize=(10, 6))
brand_counts.plot(kind='bar', color='skyblue')
plt.xlabel('EV Brand')
plt.ylabel('Count')
plt.title('Top 10 EV Brands')
plt
```

**Interpretation Of Bar chart- Top EV Manufacturers**

1️⃣ Leading Manufacturers
The manufacturers with the tallest bars represent companies that produce the largest number of electric vehicles.

2️⃣ Market Dominance
A few manufacturers usually dominate the EV market, indicating they have a strong presence in electric vehicle production.

3️⃣ Comparison Between Brands
The bar chart makes it easy to compare EV production among different manufacturers.

4️⃣ Popularity of Certain Brands
Manufacturers appearing at the top of the chart are likely more popular or widely adopted by consumers.

5️⃣ Market Concentration
If only a few companies have very high counts, it shows that the EV market is concentrated among a small number of manufacturers.


**Pie Chart - Electric Vehicle Type**

```python
#3. shows BEW vs PHEW Distribution.
vehicle_type_counts = electrical_vehicle_population['Electric Vehicle Type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(vehicle_type_counts, labels=vehicle_type_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Electric Vehicle Type Distribution')
plt.show()
```

**Interpretation Of Pie Chart - Electric Vehicle Types**


1️⃣ Proportion of Each EV Type
The pie chart shows the percentage share of BEVs and PHEVs in the dataset.

2️⃣ Dominant Vehicle Type
The larger slice of the pie indicates which EV type is more common.

3️⃣ Market Preference
If BEVs occupy a larger portion, it suggests that fully electric vehicles are more widely adopted than hybrid electric vehicles.

4️⃣ Hybrid Vehicle Presence
The PHEV portion shows the number of vehicles that use both electricity and fuel, indicating transitional adoption toward full electric mobility.

5️⃣ Overall EV Composition
The chart clearly illustrates the structure of the EV market based on vehicle type.

#2. Bivariate Analysis (5 Visuals)

**Scatter Plot - Model Year vs Electric Range**

```python
#1 .Shows how electric range changes over years
plt.figure(figsize=(10, 6))
plt.scatter(electrical_vehicle_population['Model Year'], electrical_vehicle_population['Electric Range'], alpha=0.5, color='skyblue')
```

**Interpretation Of Scatter Plot - Model Year vs Electric Range**

1️⃣ Relationship Between Year and Range
The scatter plot helps identify whether newer vehicle models have higher electric ranges compared to older models.

2️⃣ Technology Improvement
If the points trend upward as the model year increases, it indicates that EV technology has improved over time, leading to longer electric ranges.

3️⃣ Variation in Electric Range
Some points may appear spread out, showing that different manufacturers produce vehicles with different ranges even in the same year.

4️⃣ Clusters of Vehicles
Groups of points may appear in certain years, indicating periods when more electric vehicles were produced or registered.

5️⃣ Outliers
A few points may have very high or very low electric range, which could represent special models or older technology.

**Box Plot - Electric Range By Vehicle Type**

```python
#2 . Compares range differences between BEW and PHEW vehicles
sns.boxplot(x='Electric Vehicle Type', y='Electric Range',
data=electrical_vehicle_population)
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Electric Range')
plt.title('Electric Range by Vehicle Type')
```

**Interpretation Of Box Plot - Electric Range By Vehicle Type**

1️⃣ Comparison Between Vehicle Types
The box plot helps compare the electric range of BEVs and PHEVs, showing which type generally offers a higher range.

2️⃣ Median Electric Range
The line inside each box represents the median (middle value) of electric range for that vehicle type.

3️⃣ Range Distribution
The size of the box shows the spread of the data, indicating how much the electric range varies among vehicles.

4️⃣ Outliers Detection
Points outside the box represent outliers, which are vehicles with unusually high or low electric range.

5️⃣ Performance Difference
Typically, Battery Electric Vehicles (BEVs) tend to have higher electric ranges, while PHEVs usually have shorter electric-only ranges because they also use a gasoline engine.

**Bar Chart - Top Manufacturers By Average Electric Range**

```python
#3. shows the top manufacturers with the highest average electric range.
avg_range_by_manufacturer = electrical_vehicle_population.groupby('Make')['Electric Range'].mean()
top_manufacturers = avg_range_by_manufacturer.nlargest(10)
plt.figure(figsize=(10, 6))
top_manufacturers.plot(kind='bar', color='skyblue')
plt.xlabel('Manufacturer')
plt.ylabel('Average Electric Range')
plt.title('Top 10 Manufacturers by Average Electric Range')
plt.show()
```

**Interpretation Of Bar Chart - Top Manufacturers By Average Electric Range**

1️⃣ Highest Average Range Manufacturers
Manufacturers with the tallest bars have the highest average electric range, indicating stronger battery technology or high-performance EV models.

2️⃣ Comparison Between Brands
The chart makes it easy to compare the average electric range of different manufacturers and see which brands focus on longer-range vehicles.

3️⃣ Technological Advantage
Manufacturers with higher average ranges may have better battery efficiency and advanced EV technology.

4️⃣ Market Positioning
Some manufacturers may focus on premium EVs with longer ranges, while others produce vehicles with moderate ranges for affordability.

5️⃣ Consumer Insight
This chart helps identify which brands offer vehicles capable of traveling longer distances on a single charge.

**Count Plot - EV Type By Model Year**

#3. Multivariate Analysis (3 Visuals)

**Correlation Heatmap**

```python
#1 . shows the correlation between numerical variables.
# Select only numeric columns for correlation calculation
numeric_df = electrical_vehicle_population.select_dtypes(include=np.number)
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

**Interpretation Of Correlation Heatmap**

1️⃣ Correlation Strength
The heatmap displays correlation values between -1 and +1.
+1 → Strong positive correlation
0 → No relationship
-1 → Strong negative correlation

2️⃣ Positive Correlation
If two variables have a high positive value (close to +1), it means that when one variable increases, the other also increases.
Example:
Model Year and Electric Range may show positive correlation because newer EVs tend to have better range.

3️⃣ Negative Correlation
If the value is close to -1, it means when one variable increases, the other decreases.

4️⃣ Weak or No Correlation
Values close to 0 indicate little or no relationship between variables.

5️⃣ Color Representation
Warm colors (red/orange) → Strong positive correlation
Cool colors (blue) → Negative correlation
Neutral colors → Weak or no correlation.
